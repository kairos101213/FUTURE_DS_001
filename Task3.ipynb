```python
import pandas as pd
df = pd.read_csv("student_feedback.csv")
df.head()
df.columns
df.info()
```

```python

import pandas as pd

df = df.drop(columns=['Unnamed: 0'])

print("Columns after dropping:", df.columns.tolist())

print("\nColumn data types:\n", df.dtypes)

df['Average_Rating'] = df.mean(axis=1)

df.head()
```

```python
# Overall summary stats for all rating columns
df.describe()
```

```python
# Compute average per question (column-wise)
average_per_aspect = df.drop(columns=['Average_Rating']).mean().sort_values(ascending=False)
print("Average rating per aspect:")
print(average_per_aspect)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=average_per_aspect.values, y=average_per_aspect.index, palette='viridis')
plt.xlabel("Average Rating (1-5)")
plt.title("Average Ratings per Aspect")
plt.xlim(0,5)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Average_Rating'], bins=10, kde=True, color='skyblue')
plt.xlabel("Average Rating per Student")
plt.title("Distribution of Student Satisfaction")
plt.show()
```

```python
# Correlation matrix to see which aspects impact overall satisfaction most
correlation_matrix = df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Rating Aspects")
plt.show()
```

```python
#Mini Dashboard

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")


# Drop unnecessary column
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# List of numeric rating columns
rating_columns = [
    'Well versed with the subject',
    'Explains concepts in an understandable way',
    'Use of presentations',
    'Degree of difficulty of assignments',
    'Solves doubts willingly',
    'Structuring of the course',
    'Provides support for students going above and beyond',
    'Course recommendation based on relevance'
]

# Compute average rating per student using only numeric columns
df['Average_Rating'] = df[rating_columns].mean(axis=1)

# Descriptive Statistics
print("Overall descriptive statistics:\n")
print(df[rating_columns + ['Average_Rating']].describe())

# 6 Average rating per aspect
average_per_aspect = df[rating_columns].mean().sort_values(ascending=False)
print("\nAverage rating per aspect:\n", average_per_aspect)

# Pie chart - Satisfaction Buckets (fixed colors)

# satisfaction buckets
bins = [0, 2, 3, 4, 5]
labels = ['Poor (1-2)','Average (2-3)','Good (3-4)','Excellent (4-5)']
df['Satisfaction_Level'] = pd.cut(df['Average_Rating'], bins=bins, labels=labels, include_lowest=True)

# Count per bucket
satisfaction_counts = df['Satisfaction_Level'].value_counts().sort_index()

# Distinct colors for pie chart
colors = ['#fee08b', '#d73027', '#1a9850', '#4575b4']  # yellow, red, green, blue

plt.figure(figsize=(7,7))
plt.pie(satisfaction_counts, labels=satisfaction_counts.index, autopct='%1.1f%%', colors=colors)
plt.title("Student Satisfaction Levels")
plt.show()
```
